```python

import matplotlib.pyplot as plt
from sympy import lambdify, Symbol
import math
import numpy as np
```

```python
def F_x(x):
    return 5/(2+x+math.cos(x)*math.log(x + 1))
```

```python

def function(x, x_list, y_list):
    leng = len(x_list)
    summ = 0
    for i in range(leng):
        pr = 1
        for j in range(leng):
            if i !=j:
                pr *= ((x - x_list[j])/ (x_list[i] - x_list[j]))
        summ += pr*y_list[i]
    
    return summ
```

```python

diff_list = list()
    
for n in range(70):
    x_list = list()
    y_list = list()

    x_list = np.linspace(0, 10, n)
    for i in range(n):
        y_list.append(F_x(x_list[i]))
        #print (x_list[i])
    x_res = list()
    y_res = list()
    diff = 0
    for i in range(n-1):
        temp_x = (x_list[i]+x_list[i+1])/2
        x_res.append(temp_x)
        temp_y = function(temp_x, x_list, y_list)
        y_res.append(temp_y)
        if math.fabs(F_x(temp_x) - temp_y) > diff:
            diff = math.fabs(F_x(temp_x) - temp_y)
    diff_list.append(diff)
    
    if n%10 == 0:
        plt.plot(diff_list, 'r')
        plt.plot(diff_list,'b*') #
        plt.show()
        
```
