```python
#pdobn
import matplotlib.pyplot as plt
from sympy import lambdify, Symbol
import math
import numpy as np
```

```python
def F_x(x):
    return (x*x*x)/(1 - x*x*x*x)
```

```python
diff_list1 = list()#прямоугольник
diff_list2 = list()#трапеции
diff_list3 = list()#Симсон
diff_list4 = list()#Гаусс
    
for n in range(2,300):
    x_list = list()
    y_list = list()

    x_list = np.linspace(2, 4, n)
    h = 2/(n-1)
    for i in range(n):
        y_list.append(F_x(x_list[i]))
        #print (x_list[i])
    pr = 0
    tr = 0
    sim = 0
    gaus = 0
    
    for i in range(n-1):#ktdaq ghzvjeujkmybr
        pr += h*F_x((x_list[i] + x_list[i+1])/2)
    for i in range(n-1):
        tr += h*(y_list[i] + y_list[i+1])/2
    for i in range(n-1):
        sim += (h/6)*(y_list[i] + y_list[i+1] + 4*F_x((x_list[i] + x_list[i+1])/2))
        
    for i in range(n-1):
        gaus += (h/2)*(F_x((x_list[i] + x_list[i+1])/2 - (x_list[i] - x_list[i+1])/(2*math.sqrt(3))) + F_x((x_list[i] + x_list[i+1])/2 + (x_list[i] - x_list[i+1])/(2*math.sqrt(3))))
        
    diff_list1.append(math.fabs(pr - (math.log(15)- math.log(255))/4))
    diff_list2.append(math.fabs(tr - (math.log(15)- math.log(255))/4))
    diff_list3.append(math.fabs(sim - (math.log(15)- math.log(255))/4))
    diff_list4.append(math.fabs(gaus - (math.log(15)- math.log(255))/4))
    
    if n%10 == 0:
        #plt.plot(diff_list1, 'y')
        #plt.plot(diff_list1,'ys')
        #plt.plot(diff_list2, 'g')
        #plt.plot(diff_list2,'g^')
        plt.plot(diff_list3, 'b')
        plt.plot(diff_list3,'b*')
        plt.plot(diff_list4, 'r')
        plt.plot(diff_list4,'ro')
        plt.show()
```
